```python
# 9. Write a Python program to generate 1000 random numbers from a normal distribution with mean = 50 and standard deviation = 5.
# Compute its mean and standard deviation using NumPy, and draw a histogram to visualize the distribution.

import numpy as np
import matplotlib.pyplot as plt

# Parameters for the normal distribution
mean = 50
std_dev = 5
num_samples = 1000

# Generate 1000 random numbers from the normal distribution
random_numbers = np.random.normal(mean, std_dev, num_samples)

# Compute mean and standard deviation
computed_mean = np.mean(random_numbers)
computed_std_dev = np.std(random_numbers)

# Print the computed statistics
print(f"Computed Mean: {computed_mean}")
print(f"Computed Standard Deviation: {computed_std_dev}")

# Plot histogram
plt.hist(random_numbers, bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of Random Numbers from Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# 10. You are working as a data analyst for a retail company.
#The company has collected daily sales data for 2 years and wants you to identify the overall sales trend.
#daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255, 235, 260, 245, 250, 225, 270, 265, 255, 250, 260]
#Explain how you would apply the Central Limit Theorem to estimate the average sales with a 95% confidence interval.
#Write the Python code to compute the mean sales and its confidence interval.
import numpy as np
import scipy.stats as stats

daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255, 235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

# Sample statistics
mean_sales = np.mean(daily_sales)
std_dev_sales = np.std(daily_sales, ddof=1)  # Sample standard deviation
n = len(daily_sales)  # Sample size

# Confidence level and z-score for 95% confidence
confidence_level = 0.95
alpha = 1 - confidence_level
z_score = stats.norm.ppf(1 - alpha/2)

# Margin of error
margin_of_error = z_score * (std_dev_sales / np.sqrt(n))

# Confidence interval
conf_interval_lower = mean_sales - margin_of_error
conf_interval_upper = mean_sales + margin_of_error

print(f"Mean Daily Sales: {mean_sales:.2f}")
print(f"95% Confidence Interval: ({conf_interval_lower:.2f}, {conf_interval_upper:.2f})")
```
